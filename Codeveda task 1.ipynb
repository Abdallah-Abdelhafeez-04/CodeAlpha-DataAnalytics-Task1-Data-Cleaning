```python
import pandas as pd
```

```python
df= pd.read_csv('3) Sentiment dataset.csv')
df
```

```python



print("\n--- Initial DataFrame Info ---")
df.info()

print(df.head())

```

```python
if 'Unnamed: 0' in df.columns:
    df.drop('Unnamed: 0', axis=1, inplace=True)
    print("Dropped 'Unnamed: 0' column.")
```

```python


df.columns = df.columns.str.strip().str.replace(' ', '_').str.lower()
print("Standardized column names.")
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
print("Converted 'timestamp' column to datetime.")
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()
print("Stripped whitespace from string columns.")
```

```python
original_rows = len(df)
df.dropna(inplace=True)
print(f"Removed {original_rows - len(df)} rows with missing values.")
```

```python
original_rows = len(df)
df.drop_duplicates(inplace=True)
print(f"Removed {original_rows - len(df)} duplicate rows.")
```

```python
print("\n--- Final DataFrame Info After Cleaning ---")
df.info()
print("\n--- Final DataFrame Head After Cleaning ---")
print(df.head())
```
